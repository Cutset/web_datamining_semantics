## Cleaning and Converting Monuments file to JSON:

```python
df_raw = pd.read_csv("D:/ESILV_Cours/S8/Datamining_Semantics/ProjetFinal/data/liste-des-immeubles-proteges-au-titre-des-monuments-historiques.csv", sep = ';')
df_raw.head()
```

```python
df = df_raw[["appellation_courante","adresse_1","commune","coordonnees_finales","historique","reference","siecle"]]
df = df[df["commune"]=="Lyon"]

#handling coordinates:
df["lat"] = df["coordonnees_finales"].apply(lambda x : x.split(",")[0])
df["lng"] = df["coordonnees_finales"].apply(lambda x : x.split(",")[1])
df = df.drop(["coordonnees_finales"],axis=1)

#handling adresses:
for i in range(len(df)):
    try :
        df.iloc[i,1] = df.iloc[i,1].split(";")[0]
    except :
        a = 0

df = df.reset_index(drop = True)
df = df.set_axis(["nom_mon","adresse_mon","commune_mon","historique","id_mon","siecle","lat_mon","lng_mon"], axis = "columns")

#Converting centuries to int using regex:
for i in range(len(df)):
    try:
        df.iloc[i,5] = int(re.search("(\d+)(?=\s*e siècle)", df.iloc[i,5]).group(0))
    except :
        a = 0

df["type"] = "monument_historique"

df.head()
```

```python
df.to_json("D:/ESILV_Cours/S8/Datamining_Semantics/ProjetFinal/data/monuments.json", orient='records')
df.to_csv("monuments.csv")
```
